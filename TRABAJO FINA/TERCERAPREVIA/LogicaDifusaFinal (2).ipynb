```python
def recortar (y1):
    con=0
    y11=[]
    for i in y1:
        con=con+1
        if con < 12:
            y11.append(i)

    return(y11)

def recortar2 (y2):
    con=0
    y11=[]
    for i in y2:
        con=con+1
        if con > 10:
            y11.append(i)
            
    return(y11)
```
